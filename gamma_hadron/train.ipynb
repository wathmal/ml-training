```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols = ["fLength", "fWidth", "fSize", "fConc", "fConc1", "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"]
df = pd.read_csv("magic04.data", names= cols)
df.head()
```

```python
df["class"].unique()
```

```python
df.size
```

```python
print(df[df["class"] == "g"].sample(10))
df["class"] = (df["class"] == "g").astype(int)
df.head()
```

```python
for label in cols[:-1]:
    plt.hist(df[df["class"]==1][label], color='green', label='gamma', alpha=0.8, density=True)
    plt.hist(df[df["class"]==0][label], color='red', label='hadron', alpha=0.8, density=True)
    plt.title(label)
    plt.ylabel("Probability")
    plt.xlabel(label)
    plt.legend()
    plt.show()
```
